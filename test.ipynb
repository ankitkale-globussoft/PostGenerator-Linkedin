```python

import json
import pandas as pd
import re
```

```python
file_path="data/processed_posts.json"
```

```python
with open(file_path, "r", encoding="utf-8", errors="surrogatepass") as f:
    posts = json.load(f)
    
posts = json.loads(json.dumps(posts, ensure_ascii=False).encode("utf-8", "ignore").decode("utf-8"))
df = pd.json_normalize(posts)
print(df)
```
